<!-- Lọc ảnh với mean filter -->

```python
import numpy as np 
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
import colorsys
from skimage import io


a = io.imread('flower.jpeg', as_gray=True)


k = np.ones((5,5))/25

b = sn.convolve(a, k)
b_uint8 = (b * 255).clip(0,255).astype(np.uint8)
iio.imsave('flower_mean_filter.jpeg', b_uint8)

print(b_uint8)
plt.imshow(b_uint8, cmap='gray')
plt.show()
```

<!-- Lọc ảnh với median filter -->

```python
import numpy as np 
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
import colorsys
from skimage import io


a = io.imread('balloons_noisy.png', as_gray=True)
a = (a*255).astype(np.uint8)


b = sn.median_filter(a, size=5)
iio.imsave('balloons_noisy_median_filter.png',b)

print(b)
plt.imshow(b, cmap='gray')
plt.show()
```

<!-- Lọc ảnh với maximum filter -->

```python
import numpy as np 
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
import colorsys
from skimage import io


a = io.imread('baby.jpeg', as_gray=True)
a = (a*255).astype(np.uint8)


b = sn.maximum_filter(a, size=5)
iio.imsave('baby_maximum_filter.jpeg',b)

print(b)
plt.imshow(b, cmap='gray')
plt.show()
```
