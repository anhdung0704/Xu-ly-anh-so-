```python
import numpy as np 
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
import colorsys
from skimage import io


a = io.imread('dog.png', as_gray=True)


k = np.ones((5,5))/25

b = sn.convolve(a, k)
b_uint8 = (b * 255).clip(0,255).astype(np.uint8)
iio.imsave('dog_mean_filter.png', b_uint8)

print(b_uint8)
plt.imshow(b_uint8, cmap='gray')
plt.show()
```
