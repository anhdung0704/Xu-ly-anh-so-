<!-- 7) Viết chương trình sử dụng filter xác định biên của các hình trong thư mục. -->

```python
import numpy as np 
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
from skimage import filters, color

a_rgb = iio.imread('flower.jpeg')
a = color.rgb2gray(a_rgb)

b = filters.sobel(a)
b_uint8 = (b/ b.max() * 255).astype(np.uint8)

iio.imwrite ('flower_sobel_filter_edge_detection.jpeg', b_uint8)
plt.imshow(b_uint8,cmap='gray')
plt.axis('off')
plt.show()
```

<!-- Sử dụng Prewitt Filter -->

```python
import numpy as np 
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
from skimage import filters, color

a_rgb = iio.imread('flower.jpeg')
a = color.rgb2gray(a_rgb)

b = filters.prewitt(a)
b_uint8 = (b/ b.max() * 255).astype(np.uint8)

iio.imwrite ('flower_prewitt_filter_edge_detection.jpeg', b_uint8)
plt.imshow(b_uint8,cmap='gray')
plt.axis('off')
plt.show()
```
